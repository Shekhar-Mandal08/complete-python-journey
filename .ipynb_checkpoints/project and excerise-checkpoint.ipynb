```python
# - Factorial of a number 
# factorial number means product of number 5factorial = 1*2*3*4*5

num=int(input("enter a number : "))
factorial=1
for i in range(1,num+1):
    factorial=factorial*i
    
print(factorial)

# 0r

fact=1
i=1
while(i<=num):
    fact=fact*i
    i=i+1
print(fact)
```

```python
# - Accept a number and check if it a perfect number or not.
# perfect number means their factorial sum is their number equal
# eg: 5factorial =1,5 and 1 sum is 1 so no perfect
# eg: Sum of its factors (except itself) = the number
# eg:6=1,2,3,6 and 1+2+3 =6 so 6 is perfect number 


num=int(input("enter a number : "))
fact_sum=0
for i in range(1,num):
    if(num%i==0):
        print(f' {i} is this and for be perfect the sum need to be {num}')
        fact_sum=i+fact_sum

print(f'sum is : {fact_sum}') 
if fact_sum == num:

    print("it a perfect number" )
else:
    print("Not a perfect number ")

        
```

```python
# - Check wether the number is prime or not 
# prime number =2,3,5,7,11,13
# a number is divisble by itself only and 1 is not prime number got it 

num = int(input("enter a number : "))
i = 2
flag = 0       # initialize flag

while (i < num):
    if (num % i == 0):
        flag = 1
        break
    i = i + 1

if (flag == 0 and num > 1):
    print("prime number")
else:
    print("composite number")




```

```python
# - Reverse a string without using in build functions.

string=input("Enter a String you want Reverse :")
rev=string[::-1]
total_len = len(string) - 1 
for i in range(total_len,-1,-1):
    rev=string[i]
    print(rev, end=" ")
```

```python
# - Check string is Pallindrome or not 
# A palindrome is same from left to right and right to left.

string=input("enter a string  : ")
rev=""
for i  in string:
    rev=i+rev
print(rev)
if(rev==string):
    print("palindrome")
else:
    print("not palindrome")
```

```python
# - Count all letters, digits, and special symbols from a given 
# string
# i.e : number of string is how and digit and symbol
#  Given: str1 = "P@#yn26at^&i5ve"

str1 = input("enter the value,letter,symbol")

letters = 0
digits = 0
symbols = 0

for ch in str1:
    if ch.isalpha():
        letters += 1
    elif ch.isdigit():
        digits += 1
    else:
        symbols += 1

print("Letters:", letters)
print("Digits:", digits)
print("Special symbols:", symbols)

```
