```python
# 3Ô∏è‚É£ Functions & Default Arguments

# Question:
# Write a function convert_temperature(value, unit) that converts:
# Celsius ‚Üí Fahrenheit
# Fahrenheit ‚Üí Celsius
# Use default argument to specify default conversion
# Concepts: functions, default arguments

def covert(value,unit="C"):
    if unit.upper()=="C":
        return (value * 9/5)+32
    if unit.upper() =="F":
        return (value -32 )*5/9
    else:
        print("invaliid ")

print(covert(22))
print(covert(22,"F"))
```

```python
# 4Ô∏è‚É£ Lambda, Map, Filter, Reduce

# Question:
# You have a list of numbers [1, 2, 3, 4, 5].
# Filter numbers greater than 5 using filter
# Find the sum using reduce
# Concepts: functional programming, lambda

lst=[1,2,3,4,5]
# Multiply all numbers by 2 using map & lambda
print(list(map(lambda x:x*2 ,lst)))

# Filter numbers greater than 5 using filter
print(list(filter(lambda x:x>=5 ,lst)))

# Find the sum using reduce
from functools import reduce
reduce(lambda x,y:x+y,lst)

```

```python
# 7Ô∏è‚É£ File Handling

# Question:
# You have a text file sales.txt containing:
# Product1, 10
# Product2, 20
# Product1, 5
# Write Python code to:
# Calculate total quantity per product
# Save the result to a new file
# Concepts: file reading/writing, loops, string splitting


# Initialize an empty dictionary to store total quantities
totals = {}

# Open the sales.txt file in read mode
with open("sales.txt", "r") as file:
    for line in file:
        # Remove any extra whitespace and newline characters
        line = line.strip()
        # Split the line by comma
        product, quantity = line.split(",")
        # Remove extra spaces and convert quantity to integer
        product = product.strip()
        quantity = int(quantity.strip())
        
        # Add quantity to the corresponding product in dictionary
        if product in totals:
            totals[product] += quantity
        else:
            totals[product] = quantity

# Display the totals (optional)
print("Total quantities per product:", totals)

# Save the results to a new file
with open("total_sales.txt", "w") as file:
    for product, quantity in totals.items():
        file.write(f"{product}, {quantity}\n")
```
