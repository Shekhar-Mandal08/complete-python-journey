```python
class add:
    def __init__(self,a,b):
        self.result=(a+b)

a1=add(2,4)
print(a1.result)
```

```python
class Add:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def sum(self):
        return self.a + self.b

a1 = Add(2, 4)
print(a1.sum())
```
