```python
# Q4. Create a program that prints the multiplication table of a given number.
number=int(input("Enter a table Number : "))

for i in range(1,11):
    print(f"{number} * {i} = {number*i}")
```

```python
# Q5. Write a program to find the largest of three numbers.
first=float(input("enter a first number : "))
second=float(input("enter a second number : "))
third=float(input("enter a third number : "))

if first>=second and first>=third:
    print("First user is Greater ")
elif second>=first and second>=third:
    print("Second user is greater" )
else:
    print("third user is greater" )
```

```python
# Q7. Write a program to calculate the factorial of a number using a loop.
# factorial = 5!=1*2*3*4*5
# factor=5 = 1,5

num=int(input("Enter a number : "))
j=1
for i in range(1,num+1):
    j=i*j
print("Factorial Of the number is ", j)
    
    

# or for factor
num=int(input("Enter a number : "))
for z in range(1,num+1):
    if (num%z==0):
        print(z)
    
```

```python
# Q8. Create a program to count the number of vowels in a string.

string=input("enter a value : ").lower()
vowels=["a","e","i","o","u"]
print(string)
count=0
for i in string:
    if i in vowels:
        count+=1

if count==0:
    print("no any Vowels ")
else:
    print(f'count of vowels is {count}')

```

```python
# Q9. Write a Python script to reverse a given string.
string=input("enter a value : ")
rev=""
for i in string:
    rev=i+rev
print(rev)
```
