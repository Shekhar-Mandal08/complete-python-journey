```python
# ðŸ”¹ Escape Sequences in Python

txt = "We are the so-called \"Vikings\" from the north."
print(txt)
# We are the so-called "Vikings" from the north.

# 1. Single Quote (\' allows a quote inside single quotes)
print('It\'s a sunny day')  
# Output: It's a sunny day

# 2. Backslash (\\ prints a backslash)
print("This is a backslash: \\")  
# Output: This is a backslash: \

# 3. New Line (\n moves to the next line)
print("Hello\nWorld")  
# Output:
# Hello
# World

# 4. Carriage Return (\r moves cursor to start and overwrites text)
print("Hello\rWorld")  
# Output (depends on system, usually): World

# 5. Tab (\t inserts a horizontal tab space)
print("Name:\tShekhar")  
# Output: Name:    Shekhar

# 6. Backspace (\b removes the previous character)
print("Helloo\b")  
# Output: Hello

# 7. Form Feed (\f acts like a page break / moves text ahead)
print("Hello\fWorld")  
# Output (console-dependent, often like):
# Hello
#      World

# 8. Octal Value (\ooo represents characters in octal ASCII)
print("\110\145\154\154\157")  
# Output: Hello
# (Octal 110=H, 145=e, 154=l, 154=l, 157=o)

# 9. Hexadecimal Value (\xhh represents characters in hex ASCII)
print("\x48\x65\x6C\x6C\x6F")  
# Output: Hello
# (Hex 48=H, 65=e, 6C=l, 6C=l, 6F=o)
```

```python
# ðŸ”¹ Python String Methods Demonstration

text = "hello world"
text2 = "HELLO123"
space_text = "   python   "

# 1. capitalize() â†’ First letter capital
print(text.capitalize())  
# Output: Hello world

# 2. casefold() â†’ Lowercase (more aggressive than lower())
print("PYTHON".casefold())  
# Output: python

# 3. center(width) â†’ Centers string with spaces
print(text.center(20, "-"))  
# Output: ----hello world----

# 4. count(value) â†’ Count occurrences
print(text.count("l"))  
# Output: 3

# 5. encode() â†’ Encodes string (default UTF-8)
print(text.encode())  
# Output: b'hello world'

# 6. endswith(value) â†’ Check if ends
print(text.endswith("world"))  
# Output: True

# 7. expandtabs(size) â†’ Set tab size
print("H\tE\tL\tO".expandtabs(4))  
# Output: H   E   L   O

# 8. find(value) â†’ First index of substring
print(text.find("world"))  
# Output: 6

# 9. format() â†’ Format strings
print("My name is {}".format("Shekhar"))  
# Output: My name is Shekhar

# 10. format_map() â†’ Same as format but with dict
info = {"name": "Shekhar", "age": 21}
print("My name is {name}, age {age}".format_map(info))  
# Output: My name is Shekhar, age 21

# 11. index(value) â†’ Like find() but error if not found
print(text.index("world"))  
# Output: 6

# 12. isalnum() â†’ Alphanumeric only?
print("Python3".isalnum())  
# Output: True

# 13. isalpha() â†’ Alphabets only?
print("Python".isalpha())  
# Output: True

# 14. isascii() â†’ ASCII only?
print("Hello".isascii())  
# Output: True

# 15. isdecimal() â†’ Decimal only?
print("123".isdecimal())  
# Output: True

# 16. isdigit() â†’ Digits only?
print("123".isdigit())  
# Output: True

# 17. isidentifier() â†’ Valid Python variable?
print("my_var".isidentifier())  
# Output: True

# 18. islower() â†’ All lowercase?
print(text.islower())  
# Output: True

# 19. isnumeric() â†’ Numeric characters?
print("Â²345".isnumeric())  
# Output: True

# 20. isprintable() â†’ Printable chars only?
print("Hello".isprintable())  
# Output: True

# 21. isspace() â†’ Only spaces?
print("   ".isspace())  
# Output: True

# 22. istitle() â†’ Title case?
print("Hello World".istitle())  
# Output: True

# 23. isupper() â†’ All uppercase?
print(text2.isupper())  
# Output: True



# 24. join() â†’ Join iterable with separator
print("-".join(["a", "b", "c"]))  
# Output: a-b-c

# 25. ljust(width) â†’ Left align
print(text.ljust(20, "."))  
# Output: hello world.........

# 26. lower() â†’ Convert to lowercase
print("HELLO".lower())  
# Output: hello

# 27. lstrip() â†’ Remove left spaces
print(space_text.lstrip())  
# Output: 'python   '

# 28. maketrans() + translate() â†’ Replace chars
trans = str.maketrans("he", "HE")
print(text.translate(trans))  
# Output: HEllo world

# 29. partition() â†’ Split into 3 parts
print(text.partition(" "))  
# Output: ('hello', ' ', 'world')

# 30. replace() â†’ Replace substrings
print(text.replace("world", "Python"))  
# Output: hello Python

# 31. rfind() â†’ Last index of substring
print(text.rfind("l"))  
# Output: 9

# 32. rindex() â†’ Like rfind but error if not found
print(text.rindex("l"))  
# Output: 9

# 33. rjust(width) â†’ Right align
print(text.rjust(20, "."))  
# Output: .........hello world

# 34. rpartition() â†’ Split from right
print(text.rpartition(" "))  
# Output: ('hello', ' ', 'world')

# 35. rsplit() â†’ Split from right
print(text.rsplit(" ", 1))  
# Output: ['hello', 'world']

# 36. rstrip() â†’ Remove right spaces
print(space_text.rstrip())  
# Output: '   python'

# 37. split() â†’ Split into list
print(text.split(" "))  
# Output: ['hello', 'world']

# 38. splitlines() â†’ Split by line breaks
print("Hello\nWorld".splitlines())  
# Output: ['Hello', 'World']

# 39. startswith() â†’ Check if starts
print(text.startswith("hello"))  
# Output: True

# 40. strip() â†’ Remove both sides spaces
print(space_text.strip())  
# Output: 'python'

# 41. swapcase() â†’ Swap cases
print("Hello World".swapcase())  
# Output: hELLO wORLD

# 42. title() â†’ Title case
print(text.title())  
# Output: Hello World

# 43. upper() â†’ Uppercase
print(text.upper())  
# Output: HELLO WORLD

# 44. zfill(width) â†’ Pads with 0s
print("42".zfill(5))  
# Output: 00042
```

# Python Operators
### Operators are used to perform operations on variables and values.

```python
# ===============================
# PYTHON OPERATORS FULL EXAMPLES
# ===============================

# -------- Arithmetic Operators --------
x = 10
y = 3

print("Arithmetic Operators:")
print("x + y =", x + y)   # Addition â†’ 10 + 3 = 13
print("x - y =", x - y)   # Subtraction â†’ 10 - 3 = 7
print("x * y =", x * y)   # Multiplication â†’ 10 * 3 = 30
print("x / y =", x / y)   # Division (float result) â†’ 10 / 3 = 3.333...
print("x % y =", x % y)   # Modulus (remainder) â†’ 10 % 3 = 1
print("x ** y =", x ** y) # Exponentiation â†’ 10 ** 3 = 1000
print("x // y =", x // y) # Floor division â†’ 10 // 3 = 3
print()

# -------- Assignment Operators --------
print("Assignment Operators:")
a = 5
print("a =", a)           # Assign value â†’ a = 5
a += 3
print("a += 3 â†’", a)      # a = a + 3 = 8
a -= 2
print("a -= 2 â†’", a)      # a = a - 2 = 6
a *= 2
print("a *= 2 â†’", a)      # a = a * 2 = 12
a /= 3
print("a /= 3 â†’", a)      # a = a / 3 = 4.0
a %= 3
print("a %= 3 â†’", a)      # a = a % 3 = 1.0
a = 5
a //= 2
print("a //= 2 â†’", a)     # Floor divide â†’ 5 // 2 = 2
a **= 3
print("a **= 3 â†’", a)     # a = a ** 3 = 8
a = 5
a &= 3
print("a &= 3 â†’", a)      # Bitwise AND with 3 â†’ 5 & 3 = 1
a |= 3
print("a |= 3 â†’", a)      # Bitwise OR with 3 â†’ 1 | 3 = 3
a ^= 2
print("a ^= 2 â†’", a)      # Bitwise XOR â†’ 3 ^ 2 = 1
a = 8
a >>= 2
print("a >>= 2 â†’", a)     # Right shift â†’ 8 >> 2 = 2
a <<= 3
print("a <<= 3 â†’", a)     # Left shift â†’ 2 << 3 = 16
print("Walrus operator:", (b := 3))  # := assigns and returns value
print()

# -------- Comparison Operators >< --------
print("Comparison Operators:")
print("x == y â†’", x == y)   # Equal â†’ 10 == 3 = False
print("x != y â†’", x != y)   # Not equal â†’ True
print("x > y â†’", x > y)     # Greater than â†’ True
print("x < y â†’", x < y)     # Less than â†’ False
print("x >= y â†’", x >= y)   # Greater or equal â†’ True
print("x <= y â†’", x <= y)   # Less or equal â†’ False
print()

# -------- Logical Operators  and or not
# Used to combine conditional statements.--------
print("Logical Operators:")
print("x < 5 and y < 10 â†’", x < 5 and y < 10)  # Both must be True â†’ False
print("x < 15 or y < 2 â†’", x < 15 or y < 2)    # At least one True â†’ True
print("not(x < 5 and y < 10) â†’", not(x < 5 and y < 10))  # Negates result â†’ True
print()

# -------- Identity Operators  is is not 
# Identity operators are used to compare the memory location (address) of two objects, not their values.--------
print("Identity Operators:")
m = [1, 2, 3]
n = m
o = [1, 2, 3]
print("m is n â†’", m is n)         # Same object â†’ True
print("m is o â†’", m is o)         # Different objects, even if values same â†’ False
print("m is not o â†’", m is not o) # True
print()

# -------- Membership Operators in and not in 
# Used to check whether a value is present in a sequence (list, string, tuple, set, dict).--------
print("Membership Operators:")
nums = [1, 2, 3, 4, 5]
print("2 in nums â†’", 2 in nums)       # True, because 2 is in list
print("10 not in nums â†’", 10 not in nums) # True, because 10 not in list
print()

# -------- Bitwise Operators
# Work on binary (bits) of numbers.--------
print("Bitwise Operators:")
p = 6   # 110 in binary
q = 3   # 011 in binary
print("p & q =", p & q)   # AND â†’ 110 & 011 = 010 â†’ 2
print("p | q =", p | q)   # OR â†’ 110 | 011 = 111 â†’ 7
print("p ^ q =", p ^ q)   # XOR â†’ 110 ^ 011 = 101 â†’ 5
print("~p =", ~p)         # NOT â†’ inverts bits of 6 â†’ -7
print("p << 2 =", p << 2) # Left shift â†’ 110 << 2 = 11000 â†’ 24
print("p >> 2 =", p >> 2) # Right shift â†’ 110 >> 2 = 1
print()

# -------- Operator Precedence --------
print("Operator Precedence Examples:")
print("(6 + 3) - (6 + 3) =", (6 + 3) - (6 + 3))  # Parentheses first â†’ 0
print("100 + 5 * 3 =", 100 + 5 * 3)              # Multiplication before addition â†’ 115
print()
```

# List Comprehension
## List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

```python
# [expression for item in iterable if condition]
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []
# want a itesm having a 
for x in fruits:
    if "a" in x:
        newlist.append(x)
print(newlist)


# another method
newlist=[x for x in fruits if "a" in x ]
print(newlist)
    
```

```python
# newlist = [expression for item in iterable if condition == True]

newlist = [x for x in fruits if x != "apple"]
print(newlist)

newlist = [x for x in fruits]
print(newlist)

newlist = [x for x in range(10)]
print(newlist)

newlist = [x for x in range(10) if x < 5]
print(newlist)
```

# expression in list comprehension

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
# [expression for item in iterable if condition]
newlist = [x.upper() for x in fruits]
print(newlist)

newlist = ['hello' for x in fruits]
print(newlist)

newlist = [x if x != "banana" else "orange" for x in fruits]
print(newlist)

# Filter style â†’ [x for x in fruits if x != "banana"] (just removes banana)
# Ifâ€¦else expression â†’ [x if condition else y for x in fruits] (replaces banana with orange)
# if at the end â†’ filter items.
# ifâ€¦else before for â†’ replace/transform items.
```
