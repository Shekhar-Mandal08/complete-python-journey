```python
# recursion means a function call itself repeatdely   (call stack)
# print a counting number
def countingnumber(n):
    if n==0:
        return
    countingnumber(n -1)
    print(n)
    
countingnumber(5)
```

```python
# sum of two number 
def sum(a,b):
    return(a+b)
sum(4,5)
```

```python
def triange(n,i=1):
    if i>n:
        return
    print("*" * i)
    triange(n,i+1)
triange(6)
```

```python
# seee how the recursion call stack  work
def call(n):
    if n==0:
        return
    print(n)
    call(n-1)
    print("hello")
    print("call of stack is like this ")
call(3)
```

```python
# factorial number using recurion
def factorial(n):
    if (n==0 or n==1):
        return 1 
    return factorial(n-1) * n
factorial(4)
```

```python
# sum of n natural number 
def number(n):
    if (n==1):
        return 1 
    return number(n-1) + n
number(4)
```

```python
# write a recusion function to print all element  in  a lst means a lst value is in recusion
def lst(list,index=0):
    if (index==len(list)):
        return
    print(list[index])
    lst(list,index+1)
lst("334",'44','55')
    
```
