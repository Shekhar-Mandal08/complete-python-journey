```python

# but we can acces through this 

class Student:
    def __init__(self, name, age):
        self.__name = name   # private
        self.__age = age

    def display(self):
        print(f"Name: {self.__name}, Age: {self.__age}")

s1 = Student("Shekhar", 21)
s1.display()

# Direct access will fail
# print(s1.__name)  # ❌ AttributeError

# But can be accessed using name mangling:
print(s1._Student__name)  # ✅ "Shekhar"
```

```python
# inheritance(single,multi level,multiple)
# examples :
class A:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def welcome(self):
        print(f'hello welcome to my parent class {self.name} having {self.age} love to see you')


class B(A):
    pass

b1=B('shekhar',24)
print(b1.welcome())
    
```

```python
# using init constructor in child
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, rollno):  # Child made its own
        self.name = name
        self.age = age
        self.rollno = rollno


s1 = Student("Shekhar", 21, 27)
print(s1.name, s1.age, s1.rollno)
```

```python
# property decorators
class result:
    def __init__(self,phy,chem):
        self.phy=phy
        self.chem=chem
        self.percentage=str((self.phy + self.chem)/2) + "%"
r1=result(98,97)
print(r1.percentage)

# chnage the value than see
r1.phy=50
print(r1.phy)
print(r1.percentage)
```

```python

# we can seee their is no change by this so we can used a method for making it but for better is property decorator
class result:
    def __init__(self,phy,chem):
        self.phy=phy
        self.chem=chem
        # self.percentage=str((self.phy + self.chem)/2) + "%"
    def percentage(self):
         self.percentage=str((self.phy + self.chem)/2) + "%"
        
r1=result(98,97)
r1.percentage()
print(r1.percentage)
```

```python
class result:
    def __init__(self,phy,chem):
        self.phy=phy
        self.chem=chem
     # self.percentage=str((self.phy + self.chem)/2) + "%"
    @property
    def percentage(self):
         return (str((self.phy + self.chem)/2) + "%")

r1=result(98,97)
print(r1.percentage)
```
