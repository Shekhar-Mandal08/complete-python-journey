# function arguments 

```python
# postional arguments
def bio(name,age,sex):
    print(name ,'\n',age,'\n' , sex )

bio('shekhar',22,'male')
```

```python
# keywoprd arguments
def details(name,age,sex):
    print(name,age,sex)

bio(name='shekhar',sex='male',age=33)
    
```

```python
# default arguments
def details(name,sex='male'):
    print(name,sex)

details("shekhar")
details("mandal","female")
```

```python
# variable length arguments
def add(*number):
    print(number)
add(2,3,4,5,56)
```

```python
def add(** number):
    print(number)
add(name='sheekhar',age=22,sex="male")
```

```python
def total_price(*prices):
    total = 0
    for price in prices:
        total += price
    print("Total bill:", total)

total_price(100, 250, 50, 75)
```

```python
def student_info(name, *marks):
    print("Name:", name)
    print("Marks:", marks)

student_info("Shekhar", 80, 85, 90)
```

```python
def user_profile(**details):
    for key, value in details.items():
        print(key, ":", value)

user_profile(
    name="Shekhar",
    age=22,
    course="Data Science",
    hobby="Reading"
)
```

# match

```python
num = 2

match num:
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case _:
        print("Other number")
```

```python
data = [1, 2, 3]

match data:
    case [1, 2, 3]:
        print("Exact match")
    case [1, _, _]:
        print("Starts with 1")
    case _:
        print("No match")
```
